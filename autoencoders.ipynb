```python
build_df = pd.read_csv('data/post_build_encoded.csv')
build_df.info()
```

```python
build_df.shape
```

```python
y = build_df.pop('totalWBRExecutionTime')
X_train, X_test, Y_train, Y_test = train_test_split(build_df, y, test_size=0.3, random_state=0)
```

```python
X_train.head()
```

```python
N, D = build_df.shape

hidden1 = Dense(12, activation="relu")
hidden2 = Dense(6, activation="relu")
hidden3 = Dense(12, activation="relu")

stacked_encoder = Sequential([Flatten(), hidden1, hidden2])
stacked_decoder = Sequential([hidden3, Dense(N * D), Reshape([N, D])])

stacked_ae = Sequential([stacked_encoder, stacked_decoder])
stacked_ae.compile(loss="mse", optimizer="adam")
```

```python
stacked_encoder.
```

```python
history = stacked_ae.fit(X_train, X_train, epochs=100, validation_data=(X_test, X_test))
```
